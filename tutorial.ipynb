kalian bisa lihat fungsi apa aja yang ada di setiap file, udh ku buatkan daftar isinya

```python
import numpy as np  
import matplotlib.pyplot as plt
import cv2 as cv
import pandas as pd

#ini cara import setiap file dari librarynya
from pcdlib import manimg as img  
from pcdlib import manhist as hist
from pcdlib import manenhc as enhc
from pcdlib import morfimg as morf
from pcdlib import manwave as wave
```

```python
# baca gambar dulu

daun_bgr = cv.imread('dataset/Blight/Corn_Blight (263).JPG')
daun_gray = cv.cvtColor(daun_bgr, cv.COLOR_BGR2GRAY)

plt.imshow(daun_gray, cmap="gray")
```

```python
# pakai fungsi flipping dari file manimg

daun_flip = img.flipping(daun_gray, 'X')

plt.imshow(daun_flip, cmap="gray")
```

```python
# pakai fungsi wavelate dari file manwave

daun_LL, daun_LH, daun_HL, daun_HH = wave.wavelet(daun_gray)

# tampilin subbandnya juga tinggal panggil fungsi
wave.show_subbands(daun_LL, daun_LH, daun_HL, daun_HH)
```
